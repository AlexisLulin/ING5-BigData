```python
# Initialize a word counter by creating a tuple with word and cound of 1
words = rdd.flatMap(lambda lines: lines.split(' ')) \
                    .map(lambda word: (word, 1))

for w in words.take(10):
    print(w)
```

The above cell takes the rdd (full text) separate it by words in a new RDD (1 word per line as before) and then apply to the new rdd the following transformation : for each line (word) transform the line as word,1. It adds a 1 column.
This final rdd is stored in the words RDD object and is printed.

We can see in the output that all lines are composed of a word and the number 1.

```python
# a. count the occurence of each word
words_count = words.countByKey().items()
for c in list(words_count)[:10]:
    print(c)
```

```python
# b. a common first step in text analysis, change all capital letters to lower case
lower_words = words.map(lambda x: (x[0].lower(), x[1]))
for w in lower_words.take(10):
    print(w)
```

```python
# c. eliminate the stop words.
stop_words = {
    "","a", "about", "above", "after", "again", "against", "all", "am", "an",
    "and", "any", "are", "aren't", "as", "at", "be", "because", "been",
    "before", "being", "below", "between", "both", "but", "by", "can't",
    "cannot", "could", "couldn't", "did", "didn't", "do", "does", "doesn't",
    "doing", "don't", "down", "during", "each", "few", "for", "from", "further",
    "had", "hadn't", "has", "hasn't", "have", "haven't", "having", "he", "he'd",
    "he'll", "he's", "her", "here", "here's", "hers", "herself", "him", "himself",
    "his", "how", "how's", "i", "i'd", "i'll", "i'm", "i've", "if", "in", "into",
    "is", "isn't", "it", "it's", "its", "itself", "let's", "me", "more", "most",
    "mustn't", "my", "myself", "no", "nor", "not", "of", "off", "on", "once",
    "only", "or", "other", "ought", "our", "ours", "ourselves", "out", "over",
    "own", "same", "shan't", "she", "she'd", "she'll", "she's", "should",
    "shouldn't", "so", "some", "such", "than", "that", "that's", "the", "their",
    "theirs", "them", "themselves", "then", "there", "there's", "these", "they",
    "they'd", "they'll", "they're", "they've", "this", "those", "through", "to",
    "too", "under", "until", "up", "very", "was", "wasn't", "we", "we'd", "we'll",
    "we're", "we've", "were", "weren't", "what", "what's", "when", "when's",
    "where", "where's", "which", "while", "who", "who's", "whom", "why", "why's",
    "with", "won't", "would", "wouldn't", "you", "you'd", "you'll", "you're",
    "you've", "your", "yours", "yourself", "yourselves"
}

filterd_words = lower_words.filter(lambda x: x[0] not in stop_words)
for w in filterd_words.take(10):
    print(w)
```

```python
# d. sort in alphabetical order
ordered_words = filterd_words.sortByKey()
for w in ordered_words.take(5):
    print(w)
```

```python
# e. sort descending by word frequency
ordered_count = ordered_words.reduceByKey(lambda x, y: x + y).sortBy(lambda x: x[1], ascending=False)
for w in ordered_count.take(10):
    print(w)
```

```python
# f. remove punctuations and blank spaces
punctuation_list = {
    "!", '"', "#", "$", "%", "&", "'", "(", ")", "*", "+", ",", "-", ".", "/",
    ":", ";", "<", "=", ">", "?", "@", "[", "\\", "]", "^", "_", "`", "{", "|", "}", "~", "“"
}
clean_words = ordered_count.map(lambda x: (''.join(c for c in x[0] if c not in punctuation_list), x[1]))
for w in clean_words.take(10):
    print(w)

print("\n")

ordered_count2 = clean_words.reduceByKey(lambda x, y: x + y).sortBy(lambda x: x[1], ascending=False)
for w in ordered_count2.take(10):
    print(w)
```

```python
#Here is the process done at once:

final_RDD = words.map(lambda x: (x[0].lower(), x[1])).filter(lambda x: x[0] not in stop_words).map(lambda x: (''.join(c for c in x[0] if c not in punctuation_list), x[1])).reduceByKey(lambda x, y: x + y).sortBy(lambda x: x[1], ascending=False)
for w in final_RDD.take(10):
    print(w)
```

```python
 # Create an RDD of tuples (name, age)
dataRDD = sc.parallelize([("Brooke", 20), ("Denny", 31), ("Jules", 30),
("TD", 35), ("Brooke", 25)])

# Try to undestand what this code does (line by line)
agesRDD = (dataRDD
  # For each line, change the age tuple for a age and 1 count tuple the rdd is (name, (age, 1))
  .map(lambda x: (x[0], (x[1], 1)))
  # For each matching key lines (the name here) add the the ages and the 1 count
  # Result is (name, (sum of ages, count of name))
  # Each key (name) is now unique as it was reduced
  .reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1]))
  # For each line, divide the summ of the ages of the corresponding name by the number of time the name was in the original RDD giving the average age for a name
  # The new RDD is (name, average age)
  .map(lambda x: (x[0], x[1][0]/x[1][1])))
```
